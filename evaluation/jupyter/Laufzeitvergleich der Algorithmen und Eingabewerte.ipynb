```javascript
// 
var Benchmark = require('benchmark');
var suite = new Benchmark.Suite('Genetic Algorithm');

var stats;

// Create a benchmark run for each input scenario
suite.add({
    name: `Genetic ${messreihe['3feat-26m-3g']}`,
    defer: true, 
    maxTime: 60,
    fn: function(deferred) {

        // Define settings
        var settings = {
            evolutions: 1000
        }
        
        // Run async genetic algorithm
        diversity.genetic(messreihe['3feat-26m-3g'].data, messreihe['3feat-26m-3g'].groups, settings).then(result => {
            deferred.resolve();
        }).catch(err => { deferred.resolve() });
        
}}).on('complete', function () {
    console.log(this[0].stats)
    stats = this[0].stats
}).run();
```
